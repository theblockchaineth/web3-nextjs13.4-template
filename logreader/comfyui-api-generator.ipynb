```python
def queue_prompt(prompt, token):
    p = {"prompt": getPrompt(prompt, token)}
    data = json.dumps(p).encode('utf-8')
    req =  request.Request("http://127.0.0.1:8188/prompt", data=data)
    request.urlopen(req)
```

```python
## read the tokenDecisions-processed.json file and iterate through it calling queue_prompt
import json
with open('tokenDecisions-processed.json') as f:
    data = json.load(f)
    for token in data:
        queue_prompt(token['prompt'], token['tokenId'])
```
