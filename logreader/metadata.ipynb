```python
import json

## read the tokenDecisions.json file
with open('tokenDecisions.json') as json_file:
    data = json.load(json_file)

## print the data
print(data)
```

```python
base_prompt = "(perfect bold dark eyes:1.6) (sharp eyes:1.3), (detailed thick line art:1.3), 4k portrait dramatic photography, detailed, bold, High contrast, intricate by Ed Blinkey and Atey Ghailan and Jeremy Mann and Greg Rutkowski, detailed, sharp focus, dramatic, award winning, matte drawing, cinematic lighting, octane render, unreal engine, volumetrics dtx"

gender = [
    '(handsome male:1.5)', 
    '(beaufitful female:1.5)'
    ]
clique = [
    '(goth white face paint black mascara black hair, dark clothes, emo goth attire, alt rocker:1.1)',
    '(plain sports style jersey, without lettering or logos:1.1)',
    '(preppy appearance, clean cut, knitted sweater vest, khakis:1.1)',
    '(business appearance, black suit, tie, white shirt:1.1)',
    '(jeans, plain block color t-shirt without: images, brands lettering or logos:1.1)',
    '(b-boy loose fitting hip-hop urban attire with shorts:1.1)',
    '(colorful silky silk flowing flaired clothing:1.1)'
    ]
physique = [
    '(thin and skinny underweight body:1.2)',
    'average weight (perfect body:1.2)',
    'muscular (athletic body:1.2)',
    '(overweight:1.2) and obese body shape'
    ]
mentality = [
    '(in a brave pose and brave facial expression:1.2)',
    '(in a Heroic pose and Heroic facial expression:1.2)',
    '(in a Aggressive pose and Aggressive facial expression:1.2)',
    '(in a neutral pose and neutral facial expression:1.2)',
    'in a (Apathetic pose and Apathetic facial expression:1.2)',
    'in a (Nervous pose and Nervous facial expression:1.2)',
    'in a (Scared pose and Scared facial expression:1.2)'
    ]

hair = [
    '(short hair:1.1)',
    '(long hair:1.1)',
    '(curly hair:1.1)',
    '(straight hair:1.1)',
    '(wavy hair:1.1)',
    '(bald hair:1.1)'
    ]
background_location = [
    '(city waterfront harbour:1.2)',
    '(city market square:1.2)',
    '(dimly lit city alleyway:1.2)',
    '(city police station headquarter:1.2)',
    'outside of a (abandoned warehouse outskirts of the city with skyscrapers in the background:1.2)',
    '(interior shot of a smoky bar pub:1.2)',
    '(exterior shot of a vintage american diner:1.2)',
    'in a (city park:1.2)',
    'outside of an (industrial complex:1.2)',
    '(exterior shot of a creepy mansion within the city with skyscrapers in the background:1.2)',
    '(foggy city streets:1.2)',
    '(exterior shot of a cemetery:1.2)'
    ]
time = [
    '(golden hour time of day:1.2)', 
    '(blue hour time of day:1.2)', 
    '(midnight time of day:1.2)', 
    '(afternoon time of day:1.2)', 
    '(dusk time of day:1.2)'
    ]
lighting = [
    '(backlit:1.2)', 
    '(backlit:1.2)', 
    '(backlit:1.2)', 
    '(high contrast dramatic side lit lighting:1.2)'
    ]
race = [
    '(causian:1.1)', 
    '(causian:1.1)', 
    '(causian:1.1)', 
    '(african american:1.1)', 
    '(spanish hispanic:1.1)', 
    ]
```

```python
## for each token, print the token and the decision
import random

def fn_GetPrompt(array, idx):
    print(array)
    print(idx)
    if idx == 0:
        print('random')
        choice = random.randrange(1, len(array)-1)
        return array[choice]
    else:
        print('found')
        return array[idx]

prompt_outputs = []
for token in data:
    gender_selected = token['attributes'][2]['code']
    clique_selected = token['attributes'][3]['code']
    physique_selected = token['attributes'][4]['code']
    mentality_selected = token['attributes'][5]['code']

    trait_list = []
    trait_list.append(random.choice(race))

    if gender_selected == 0:
        trait_list.append(random.choice(gender))
    else:
        trait_list.append(gender[gender_selected - 1])

    if clique_selected == 0:
        trait_list.append(random.choice(clique))
    else:
        trait_list.append(clique[clique_selected - 1])

    trait_list.append(random.choice(hair))

    if physique_selected == 0:
        trait_list.append(random.choice(physique))
    else:
        trait_list.append(physique[physique_selected - 1])

    if mentality_selected == 0:
        trait_list.append(random.choice(mentality))
    else:
        trait_list.append(mentality[mentality_selected - 1])

    trait_list.append(random.choice(background_location))
    trait_list.append(random.choice(time))
    trait_list.append(random.choice(lighting))

    prompt = ', '.join(trait_list) + ', ' + base_prompt

    print(prompt)
    token['prompt'] = prompt
    prompt_outputs.append(token)


## write the tokenDecisions-processed.json file
with open('tokenDecisions-processed.json', 'w') as outfile:
    json.dump(prompt_outputs, outfile)
```
