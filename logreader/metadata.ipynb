```python
## for each token, print the token and the decision
import random

def fn_GetPrompt(array, idx):
    print(array)
    print(idx)
    if idx == 0:
        print('random')
        choice = random.randrange(1, len(array)-1)
        return array[choice]
    else:
        print('found')
        return array[idx]

prompt_outputs = []
for token in data:
    gender_selected = token['attributes'][2]['code']
    clique_selected = token['attributes'][3]['code']
    physique_selected = token['attributes'][4]['code']
    mentality_selected = token['attributes'][5]['code']

    trait_list = []
    trait_list.append(random.choice(race))

    if gender_selected == 0:
        trait_list.append(random.choice(gender))
    else:
        trait_list.append(gender[gender_selected - 1])

    trait_list.append(random.choice(['25 years old', '35 years old', '25 years old', '35 years old', '25 years old', '35 years old', '25 years old', '35 years old', '45 years old', '65 years old',]))

    if gender_selected == 1:
        trait_list.append(random.choice(['clean shaven', 'clean shaven', 'clean shaven', 'clean shaven', 'clean shaven', 'clean shaven', 'bearded', 'goatee', 'mustache', 'sideburns', 'stubble']))

    if gender_selected == 2:
        trait_list.append(random.choice(['bold makeup', 'lipstick', 'dark makeup', 'eyeliner eye liner', '', '', '', '']))

    if clique_selected == 0:
        trait_list.append(random.choice(clique))
    else:
        trait_list.append(clique[clique_selected - 1])

    trait_list.append(random.choice(hair))

    if physique_selected == 0:
        trait_list.append(random.choice(physique))
    else:
        trait_list.append(physique[physique_selected - 1])

    if mentality_selected == 0:
        trait_list.append(random.choice(mentality))
    else:
        trait_list.append(mentality[mentality_selected - 1])

    trait_list.append(random.choice(background_location))
    trait_list.append(random.choice(time))
    trait_list.append(random.choice(lighting))

    prompt = 'mid-distance photograph ' + ', '.join(trait_list) + ', ' + base_prompt

    print(prompt)
    token['prompt'] = prompt
    prompt_outputs.append(token)


## write the tokenDecisions-processed.json file
with open('tokenDecisions-processed.json', 'w') as outfile:
    json.dump(prompt_outputs, outfile)
```
